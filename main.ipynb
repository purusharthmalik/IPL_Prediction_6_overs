```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
#Loading data
import glob

df = pd.DataFrame()

for _ in glob.glob("*.csv"):
    df = df.append(pd.read_csv(_))
```

```python
#making a temp data
df1 = df.copy()
#Dropping duplicate rows
df1.drop_duplicates(subset = ['start_date', 'batting_team', 'bowling_team', 'innings', 'ball'], inplace = True)
```

```python
#Resetting the index
df1.reset_index(inplace = True)
#Dropping and index
df1.drop("index", axis = 1, inplace  = True)
```

```python
#Setting column limit to None
pd.set_option("display.max_columns", None)
```

```python
#Changing dtype of date
df1.start_date = pd.to_datetime(df1.start_date)
```

```python
#getting the season right
for (i, j) in zip(df1.index, df1.season):
    col = df1.columns.get_loc("season")
    if type(j) == int:
        pass
    elif type(j) == str:
        df1.iloc[i, col] = df1.iloc[i, df1.columns.get_loc("start_date")].year
    else:
        df1.iloc[i, col] = np.nan
```

```python
#Changing the dtype of season
df1.season = df1.season.astype(int)
```

```python
#Adding dropping start_date
df1.drop("start_date", axis = 1, inplace = True)
```

```python
#Changing NaN values to 0
for _ in ['wides', 'noballs', 'byes', 'legbyes', 'penalty']:
    df1[_] = df1[_].apply(lambda x: int(0) if math.isnan(x) else int(x))
```

```python
df1.head()
```

```python
df1['total_runs'] = df1.runs_off_bat + df1.extras
```

```python
df1.groupby(['match_id', 'innings']).sum()
```

```python
l = [0, 1, 2, 3, 4, 5, 6]
b = []
for i in l[:-1]:
    for j in l[1:]:
        b.append(f'{i}.{j}')
b = list(map(float, b))
```

```python
df2 = df1[df1['ball'].apply(lambda x: x in b)]
```

```python
df2.drop(['extras', 'non_striker', 'wicket_type', 'player_dismissed', 'season', 'wides', 'noballs', 'byes', 'legbyes', 'penalty', 'other_wicket_type', 'other_player_dismissed'], axis = 1, inplace = True)
```

```python
batsmen = df2.groupby(['match_id', 'innings'])['striker'].apply(lambda x: list(set('\t'.join(x).split('\t')))).reset_index()
bowlers = df2.groupby(['match_id', 'innings'])['bowler'].apply(lambda x: list(set('\t'.join(x).split('\t')))).reset_index()
bat_team = df2.groupby(['match_id', 'innings'])['batting_team'].apply(lambda x: list(set('\t'.join(x).split('\t')))[0]).reset_index()
bowl_team = df2.groupby(['match_id', 'innings'])['bowling_team'].apply(lambda x: list(set('\t'.join(x).split('\t')))[0]).reset_index()
venue = df2.groupby(['match_id', 'innings'])['venue'].apply(lambda x: list(set('\t'.join(x).split('\t')))[0]).reset_index()
total_runs = df2.groupby(['match_id', 'innings']).sum()
```

```python
final_data = pd.DataFrame([venue.venue, venue.innings, bat_team.batting_team, bowl_team.bowling_team, batsmen.striker, bowlers.bowler, total_runs.total_runs.values])
```

```python
final_data = final_data.transpose()
```

```python
final_data.rename(columns = {'Unnamed 0':'total_runs'}, inplace = True)
```

```python
final_data.head()
```

```python
final_data.to_csv('final_data.csv', index=False)
```

```python
data.head()
```

```python
def get_firsts(x):
    return ''.join([_[0] for _ in x.split()])
```

```python
data.venue = data.venue.apply(lambda x: get_firsts(x))
data.batting_team = data.batting_team.apply(lambda x: get_firsts(x))
data.bowling_team = data.bowling_team.apply(lambda x: get_firsts(x))
```

```python
data.innings = data.innings.apply(lambda x: 0 if x == 1 else 1)
```

```python
data.striker = data.striker.apply(lambda x: x.replace(" ", '').replace("-", ''))
```

```python
data.bowler = data.bowler.apply(lambda x: x.replace(" ", '').replace("-", '').replace('(2)', ''))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(min_df=0, lowercase=False)
vectorizer.fit(data.batting_team)
teams = vectorizer.vocabulary_
```

```python
vec_new = CountVectorizer(min_df = 0, lowercase = False)
vec_new.fit(data.venue)
v = vec_new.vocabulary_
v['N'] = 38
v['K'] = 39
```

```python
vec_striker = CountVectorizer(min_df = 0, lowercase = False)
vec_striker.fit(data.striker)
v_strike = vec_striker.vocabulary_
```

```python
vec_bowler = CountVectorizer(min_df = 0, lowercase = False)
vec_bowler.fit(data.bowler)
v_bowl = vec_bowler.vocabulary_
```
